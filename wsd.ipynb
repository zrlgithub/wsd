Formare lista de cuvinte din fisierul stopwords.txt (ex. prepozitii si cuvinte care nu exprima ceva ce poate face diferenta intre cele doua definitii)

```python
lines = []
with open('drive/MyDrive/Colab Notebooks/stopwords.txt') as f:
    lines = [line.rstrip('\n') for line in f]

print(lines)
```

```python
import rowordnet as rwn
wordnet = rwn.RoWordNet()

def overlapcontext( synset, sentence ):
    gloss = wordnet.synset(synset).definition.split(" ")
    gloss=set(gloss).union()-set(lines).union() #scoatem cuvintele din stopwords.txt(1)
    if isinstance(sentence, str):
        sentence = set(sentence.split(" "))
    elif isinstance(sentence, list):
        sentence = set(sentence)
    elif isinstance(sentence, set):
        pass
    else:
        return
    sentence = set(sentence).union()-set(lines).union() #(1)
    gloss=set(gloss)
    sentence=set(sentence)
    return len( gloss.intersection(sentence) ) #numarul de cuvinte egale

def lesk( word, sentence ):
    bestsense = None
    maxoverlap = 0
    word=wordnet.synsets(word) if wordnet.synsets(word) is not None else word
    for sense in word: #word devine lista de synset-uri a cuvantului in sine prin rowordnet
        overlap = overlapcontext(sense,sentence)
        if overlap > maxoverlap: #verificare synset
                maxoverlap = overlap
                bestsense = sense
    return bestsense

count=0 #numar total verificari corecte
counter=0 #numar verificari corecte pe fiecare cuvant
all=0 #numar total 
for column in df.columns:
  for row in df[column]:
      counter=0
      for synset in row:
        a = lesk(column,synset["sentence"]) #synset gasit
        all=all+1
        if(a == synset['correct_synset_id']): #verificare cu cel bun din dataset
            count=count+1
            counter=counter+1
  print("Nr of correct synsets for word: " ,column,counter)
print("\n\n Total number of correct synsets=",count)
print("\nTotal number of verification=",all)
print("\nPrecision=",100*count/all,"%")
```
